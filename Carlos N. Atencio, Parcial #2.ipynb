```python
A=8.997
B=93.7967
C=-53.12099
D=12.1159
E=0.36272

T=np.arange(298,1200,1)      # Rango de temperatura indicado por el profesor
dT=1                         # Cambio de temperatura
t=T/1000
Cp=A+B*t+C*t**2+D*t**3+E/t**2
Cp
```

```python
##valores de Delta H y Delta S para el calculo del Delta G

H=np.cumsum(Cp*dT)+55.13*1000
H[-1]
S=np.cumsum(Cp/T*dT)
G=H-T*S
G
```

```python
plt.plot(T,G,color="red")
plt.title("Delta G vs T")
plt.xlabel("T (K)")
plt.ylabel(r'$\Delta G (J/mol)$')
plt.show()
```

```python
###Problema #2  Calcule la constante de equilibrio para el sistema:
```

```python
deltaS_NO=210.758
deltaH_NO=90.291
deltaS_O2=161.058
deltaH_O2=249.173
deltaS_NO2=240.034
deltaH_NO2=33.095

T=298.15     #Temp=Kelvin
pt=1         #Presion=Bar
R=8.314 
```

```python
##Calculamos la energia libre estandar
uNO_est=deltaH_NO-T*deltaS_NO
uO2_est=deltaH_O2-T*deltaS_O2
uNO2_est=deltaH_NO2-T*deltaS_NO2
print(uNO_est,uO2_est,uNO2_est)
```

```python
### Potencial químico en función de T,p y Bx
x=np.arange(0.001,1,0.001)
n_NO=1-x
n_NO2=x
n_O2=(1/2)-(1/2)*x
nt=(3/2)-(1/2)*x

X_NO=n_NO/nt
X_NO2=n_NO2/nt
X_O2=n_O2/nt
```

```python
#xi=ξ en las siguientes fórmulas

uNO=uNO_est+R*T*np.log(X_NO)
uNO2=uNO2_est+R*T*np.log(X_NO2)
uO2=uO2_est+R*T*np.log(X_O2)
uT=n_NO*uNO+n_NO2*uNO2+n_O2*uO2
```

```python
minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)


plt.plot(x,uT,color="red")
plt.axvline(x=xeq,color="green",linestyle="-.")
plt.xlabel(r"$\xi$")
plt.ylabel(r"$\Delta G$ (J/mol)")
plt.show()
```

```python
pNO2=((xeq)/((3/2)-(1/2)*xeq))*pt
pNO=(1-xeq/((3/2)-(1/2)*xeq))*pt
pO2=(((1/2)-(1/2)*xeq)/((3/2)-(1/2)*xeq))*pt

Kp=((pNO2)/(pNO*pO2**(1/2)))
print(Kp)
```

```python
def secuenciaADN(i):
    a = np.arange(4)
    cad=(np.random.choice(a, i, replace=True))
    sec=str(cad)
    sec1=sec.replace('0','A')
    sec2=sec1.replace('1','T')
    sec3=sec2.replace('2','C')
    sec4=sec3.replace('3','G')
    return(sec4)
```

```python
x=secuenciaADN(1000)
y=x[1]
```
