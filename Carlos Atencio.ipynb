```python
formula="C3H5N3O9" ##nitroglicerina

elem=[]
cad=""
ca=""
cad=""
for x in formula:
    if x>="A" and x<="Z":
        elem.append(x)
    if x>="a" and x<="z":
        s=elem.pop()
        s=s+x
        elem.append(s)
        
    if (x>="0" and x<="9") and (ca=="LM" or ca=="lm"):
        cad=cad+"*"
    if x=="(" and (ca=="N" or ca=="LM" or ca=="lm"):
        cad=cad+"+"
    if x>="0" and x<="9" and ca==")":
        cad=cad+"*"
    if (x>="A" and x<="Z") and (ca=="LM" or ca=="lm"):
            cad= cad+"+"
    if (x>="A" and x<="Z") and ca=="N":
        cad=cad+"+"
        
    cad=cad+x
    
    if x>="A" and x<="Z":
        ca="LM"
    if x>="a" and x<="z":
        ca="Lm"
    if x>="0" and x<="9":
        ca="N"
    if x=="(":
        ca="("
    if x==")":
        ca=")"
        
print(cad)
for e in elem:
    peso=str(elementos[e][0])
    cad=cad.replace(e,peso)
PM=eval(cad)
print(PM)
```

```python
x=np.arange(10)  
A=3.52
B=-2.85
C=5.16
y=((A)+((x)**10*B)+(C*x**2))/2
plt.plot(x,y)
plt.xlabel("Valores de 1 al 10")
plt.ylabel("Eje de las Y")
plt.title("Funcion Y=A+log(x)*B+C*x^2/2")
plt.show
```
