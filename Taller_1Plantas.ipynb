```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe["target"]=iris.target
x=iris_dataframe["SL"]
print(x.mean())
y=iris_dataframe["SW"]
col=(iris_dataframe["target"]+1)*2
print(y.mean())
PL=iris_dataframe["PL"]*50
PW=iris_dataframe["PW"]*100
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=0.5)
plt.xlabel("Largo del spalo (cm)")
plt.ylabel("Ancho del Spalo (cm)")
plt.title("Ancho de Spalo vs Largo del Spalo")
plt.grid(True)
plt.show()
```
