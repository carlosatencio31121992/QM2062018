```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fraccion(pH,pka):
    H=10**(-pH)
    ka=[1]
    for pk in pka:
        x=10**(-pk)
        ka.append(x)
    n=len(ka)-1
    mult=1
    terminos=[]
    for a in range(n+1):
        n1=(n-a)
        cH=H**n1
        mult=mult*ka[a]
        d=cH*mult
        terminos.append(d)
    sumaTerminos=sum(terminos)
    alpha=terminos/sumaTerminos
    return alpha
```

```python
pka=np.array([4.0,6.7,8.5])
pH=0
x=[]
y1=[]
y2=[]
y3=[]
y4=[]
while pH<=14:    
    x.append(pH)
    alpha=fraccion(pH,pka)
    y1.append(alpha[0])
    y2.append(alpha[1])
    y3.append(alpha[2])
    y4.append(alpha[3])
    pH+=0.1
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.xlabel("pH")
plt.ylabel("francion de las especies")
plt.title("F.Esp. Ac.")
plt.show()
```
